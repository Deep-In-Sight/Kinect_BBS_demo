```python
from glob import glob 
import os
import shutil
```

```python
root = "/home/eckim/workspace/fhe_etri/Data/BBS/Sample_BBS/BBS"
```

```python
id_list = glob(root+'/e/*/*/*/*/*/*/*.jpg')
id_list.sort()
```

```python
for i in id_list:
    file = i.split(root)[1]
    _, cam, ID, _, _, _, class_name, score, jpg = file.split("/")
    new_root = (f'{root}/frame_1_idx_e/{class_name}/{score}/{ID}/')
    os.makedirs(new_root, exist_ok=True)
    shutil.move(i, new_root+jpg)
    print(new_root+jpg)
```
