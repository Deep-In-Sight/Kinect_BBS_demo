```python
import numpy as np
import re

def remove_special_character(s):
    return re.sub(r'\W+', '', s)

def shorten_name(name):
    name = remove_special_character(name)
    name = name.replace('left', 'L')
    name = name.replace('right', 'R')
    return name

def convert_to_csv(fn):
    out_fn = fn.replace("/","_")
    #out_fn = "/home/hoseung/Work/data/BBS/dense/"+out_fn

    with open(fn) as f:
        err_at_end=-1
        all_lines = f.readlines()
        for i in range(len(all_lines)):
            try:
                last_line = all_lines[err_at_end]
                #print(last_line)
                last_frame = int(last_line.split(",\"")[0])
                break
            except:
                err_at_end -=1
        if i == len(all_lines) -1:
            print("Cannot find a valid line")
            return 
        nframes = last_frame +1

        # npersons
        npersons = len(last_line.split(",\""))-1
        print(last_line.split(",\""))
        print("Nperson", npersons)
        # Get names 
        ll = last_line.split(",\"")[1] # [0] = frame count
        names = []
        for lll in ll.split("],"):
            llll = lll.split(",")
            name = shorten_name(llll[0])
            names.append(name)

    prefix_dims = ["x_", "y_"]
    dtypes = [(pref + name, int) for name in names for pref in prefix_dims]

    njoints = len(names)
    
    skels = np.zeros((nframes, npersons),
                    dtype=dtypes)

    with open(fn) as f:
        f.readline()
        for iframe, l in enumerate(f.readlines()):
            try:
                print(l.split(",\"")[1:])
                print("N split", len(l.split(",\"")[1:]))
                print("np", npersons)
                for ip, ll in enumerate(l.split(",\"")[1:]):
                    for lll in ll.split("],"):
                        llll = lll.split(",")
                        name = shorten_name(llll[0])
                        skels['x_'+name][iframe, ip] = int(remove_special_character(llll[1]))
                        skels['y_'+name][iframe, ip] = int(remove_special_character(llll[2]))
            except:
                print(fn, "error in this line", l[:10])

    for i in range(npersons):
        df = pd.DataFrame(skels[:,i])
        df.to_csv(out_fn.replace("bodytracking_data", f"bt_person{i}"))
```
